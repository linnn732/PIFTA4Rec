```python
import pandas as pd
from collections import defaultdict

import pickle
import os
import pandas as pd
import numpy as np
import random
from tqdm import tqdm
```

```python
#DATASET_NAME = "TaFeng"
DATASET_NAME = "Dunnhumby"
```

#### TaFeng

```python
# 讀取資料集
df = pd.read_csv(f'../../cleaned_dataset/{DATASET_NAME}_clean.csv')
df = df[['CUSTOMER_ID','CART_ID','NEW_ITEM_ID']]
df
```

```python
userid_dict = {}
new_id = 0
def give_user_id(x):
    if x not in userid_dict.keys():
        if userid_dict:
            new_id = max(userid_dict.values()) + 1
        else:
            new_id = 0
        
        userid_dict[x] = new_id
        
    return userid_dict[x]

tqdm.pandas(desc = 'apply')
new_userID_series = df["CUSTOMER_ID"].progress_apply( give_user_id )
df["CUSTOMER_ID"] = new_userID_series
df
```

```python
# 選擇需要使用的欄位
df = df[['CUSTOMER_ID', 'CART_ID', 'NEW_ITEM_ID']]

# 以 CUSTOMER_ID 為基準進行分組
grouped = df.groupby('CUSTOMER_ID')

# 轉換資料格式
result = []
for name, group in grouped:
    # 將每個 CART_ID 視為一個會話，將 CART_ID 中的 new_item_id 作為該會話的項目
    sessions = group.groupby('CART_ID')['NEW_ITEM_ID'].apply(list).tolist()
    # 將每個使用者的會話轉換成一個字串
    session_strs = [':'.join([str(item) for item in session]) for session in sessions]
    # 將使用者 ID 和該字串作為一行輸出
    result.append([name, '@'.join(session_strs)])

# 將結果轉換為 DataFrame
result_df = pd.DataFrame(result, columns=['user', 'sessions'])
result_df
```

```python
# 計算用戶總數和項目總數
user_count = df['CUSTOMER_ID'].nunique()
item_count = df['NEW_ITEM_ID'].nunique()
print(user_count)
item_count
```

```python
# 插入用戶總數和項目總數到 DataFrame 的第一行
result_df.loc[-1] = [user_count, item_count]  # -1 代表插入到第一行
result_df.index += 1  # 調整所有索引以避免重複的行索引
train_data = result_df.sort_index()  # 重新排序 DataFrame 的索引
train_data
```

```python
# 儲存 DataFrame 為 CSV 檔案
preprocessing_folder = "data"
if not os.path.exists(preprocessing_folder):
    os.mkdir(preprocessing_folder)
train_data.to_csv(f'data/{DATASET_NAME}.csv', index=False, header=None)
```
